```python
#Ejercicio 6 computacional
import numpy as np
import matplotlib.pyplot as plt

# Definimos g(x)
def g(x):
    return np.pi + 0.5 * np.sin(x / 2)

# Método de punto fijo con 4 iteraciones reales
def punto_fijo(g, x0, tol=1e-2, iteraciones=4):
    print(f"Valor inicial: x0 = {x0:.6f}")
    for i in range(1, iteraciones + 1):  # 4 iteraciones reales
        x1 = g(x0)
        print(f"Iteración {i}: x{i} = {x1:.6f}")
        if abs(x1 - x0) < tol:
            print(f"\nSe alcanzó la tolerancia en la iteración {i}. Aproximación final: {x1:.6f}")
            break
        x0 = x1
    else:
        print(f"\nAproximación después de {iteraciones} iteraciones: x ≈ {x0:.6f}")

# Ejecutar
x0 = 2.0
punto_fijo(g, x0)

# Gráfica
x_vals = np.linspace(0, 2 * np.pi, 400)
g_vals = g(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, g_vals, label=r'$g(x) = \pi + 0.5\sin(x/2)$', color='blue')
plt.plot(x_vals, x_vals, label=r'$y = x$', color='red', linestyle='--')
plt.title('Gráfica de $g(x)$ y la identidad $y = x$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```
